```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
#using tips dataset
# Loading data set 
tips = sns.load_dataset('tips')
```

```python
#concise summary of the data set.
tips.info()
```

```python
tips.head()
```

```python
#The dataset consists f data pertaining to customers,
#The data has a total_bill along with the amount of tip the customer left for the staff along with some other information such as day, sex, time, size of the group/party etc
```

```python
#distplots for univariate analysis
# lets check how the total bill are distributed in the data set
sns.distplot(tips['total_bill'],color='Orange')
```

```python
#kde is a more smooth curve and better than hist graph which vary by the bins chosen
#we can see common occurance of total bill given on any day is 12-25$
```

```python
sns.distplot(tips['tip'],color= 'blue')
```

```python
#similarly common tip is around 2-4$
```

```python
#more interactive plot is jointplot
sns.jointplot(x='total_bill',y='tip',data=tips)#kind='scatter' - default
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #hexagon plot
```

```python
#This will give 2D kde plot, 
#showing the darker region with density where most of the data points match up.
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#pairplot 
sns.pairplot(tips)
```

```python
#we can provide hue aswell to get caegorical plot
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#passing total_bill to the rugplot which 
#plot sticks on an axis for every data point on a univariate distribution
sns.rugplot(tips['total_bill'])
```

```python
#combining rug plot and kde plot is very helpful
sns.rugplot(tips['total_bill'],color='r')
sns.distplot(tips['total_bill'], kde=True)
#we can see from below plot that the intensity of red lines in rug plot is same as kde plot distribution
```

```python
#Categorical scatterplots
sns.stripplot(x="day", y="total_bill",data=tips)
```

```python
#In the above strip plot, we see that the datapoints are overlap, 
# and makes it difficult to observe the full distribution of data
sns.stripplot(x="day", y="total_bill",data=tips,
              jitter=True,
              hue="sex")
```

```python
#swarm plot: better representation of overlappng data
sns.swarmplot(x="day", y="total_bill",data=tips,
              hue="sex")#,palette='coolwarm')
```

```python
#On Fridays, not many go out for lunch or dinner as compared to weekends
#A promotional team would look at this plot and plan to provide some offers for friday to boost sales
```

```python
#A more thorough indepth understanding of a category can be achieved with boxplotting 
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
#The mean for thursday is 15, for friday also around the same but for sat and sun it is 
#at a little higher value of 20$. The min and max of boxplot can help us understand the range of bills by day 
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,
           hue = 'sex')
```

```python
#On any given day, males average tend to stay high .
#Interestingly on a saturday they share similar distrbution showing female prefer to plan lunch or dinners on saturdays  
```

```python
#the violin plot allows us to actually plot all the components that corresponds to actual data points. 
#The violin plot is essentially showing the kernel density estimation of the underlying distribution
```

```python
sns.violinplot(x="day", y="total_bill",data=tips,
               hue='sex', split=True)
```

```python
#sns.barplot(x='sex', y='total_bill',data=tips) #estimator is mean here

sns.barplot(x="day", y="total_bill", hue="sex",data=tips)
```

```python
#sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
sns.barplot(x="day", y="total_bill", data=tips, hue="sex",estimator=np.std)
```

```python
# a simple count plot 
sns.countplot(x="day", hue="sex",data=tips)
```

```python
# another example using size in the data
sns.countplot(x="size",data=tips)
# group of 2 is the most common!
```

```python
# combining barplot on top of pointplot to compare
sns.pointplot(x="day", y="total_bill", hue="sex",data=tips)
sns.barplot(x="day", y="total_bill", hue="sex",data=tips, alpha=0.5)
plt.ylim(0,30);
```

```python
# How much in tip on each day
sns.violinplot(x="day", y="tip", data=tips, palette="rainbow")
sns.swarmplot(x="day", y="tip", data=tips, color="black",size=3)
```

```python
#fcatorplot is a general categorical plot that takes bar, violin, swarm as a kind
sns.factorplot(x="sex",y="total_bill",data=tips,
               hue='smoker',kind="box")
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker",
               col="time", data=tips, kind="swarm")
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker",
               col="day", data=tips, kind="box",size=4, aspect=.5)
```
